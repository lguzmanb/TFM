```python

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

from keras.optimizers import SGD
from sklearn.model_selection import StratifiedKFold
from scipy.stats import uniform as sp_rand
from scipy.stats import randint
from time import time
from sklearn import preprocessing

import math
from sklearn.metrics import mean_squared_error
```

```python
# definim model
import tensorflow as tf
model =Sequential()
model.add(LSTM(50, activation='relu', input_shape=(48, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
#normalitzem
scaler=preprocessing.MinMaxScaler(feature_range=(0, 1))
hourly_norm=scaler.fit_transform(hourly_h)
```

```python
#Seleccionem dades per y i X
y_hourly=hourly_norm[:,0]
X_hourly=hourly_norm[:,1:49]

#Reshape de [samples,timesteps] a [samples,timesteps,features]

X_hourly=np.reshape(X_hourly, (X_hourly.shape[0],48,1))
```

```python
len(X_hourly)
```

```python
#Walk forward per test i train
minim=5000
n_train=5336#Utilitzem 14 dies de dades horaries que son 336 hores
lenght=5672

llista_prediccions=list()
llista_preditrain=list()
real=list()

for i in range(n_train,lenght):
    minim=minim+1
    X_train,X_test= X_hourly[minim:i],X_hourly[i:i+1]
    y_train,y_test= y_hourly[minim:i],y_hourly[i:i+1]

    #fem fit al model
    model.fit(X_train, y_train, epochs=50, verbose=0)
    
    #Predim per cadascun
    preditest=model.predict(X_test)
    llista_prediccions.append(preditest)
    
    preditrain=model.predict(X_train)
    llista_preditrain.append(preditrain)
    
    real.append(y_test)
    
```

```python
testScore=list()
llistaerror=list()
for i in range(len(llista_prediccions)):
    #Calculem RMSE
    testScore.append(math.sqrt(mean_squared_error(real[i], llista_prediccions[i])))
    
    #Calculem error
    error= math.fabs(real[i]-llista_prediccions[i])/real[i]
    llistaerror.append(error)
    
    
    
```
